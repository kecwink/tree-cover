```python
#print list of tree columns

tree_df.columns

# #drop columns
tree_inv = tree_df[['SITE_ID', 'SPECIES_BOTANIC', 'DIAMETER','NEIGHBOR', 'X_LONG', 'Y_LAT']]
tree_inv

#rename columns
tree_inventory = tree_inv.rename(columns={"SITE_ID": "Tree_ID",
                                          'SPECIES_BOTANIC':'Tree_Species',
                                          'DIAMETER':'Tree_Diameter',
                                          'NEIGHBOR':'Neighborhood',
                                            'X_LONG': "Longitude",
                                         'Y_LAT': 'Latitude'})
tree_inventory

#check for blanks
tree_inventory.count()



#duplicate/similar neighborhood names?

tree_inventory.columns
neighborhoods = tree_inventory['Neighborhood'].unique()
neighborhoods



```

```python
#merge

merge_df = demographics.merge(tree_inventory, on='Neighborhood')
merge_df

#change index
merge_df.set_index('Neighborhood_id', inplace=True)
merge_df


merge_df
```

```python
#count the number of trees in each neighborhood
trees_per_neighborhood = merge_df.groupby('Neighborhood_id').count()['Tree_ID']
#type(trees_per_neighborhood)
# trees =[]
#  for i in range(len(trees_per_neighborhood)):
#      trees.append()
trees_per_neighborhood
```

```python
# isolate neighborhood names
neighborhoods = merge_df.groupby('Neighborhood_id').first()
n_names = []
for i in range(len(neighborhoods)):
    n_names.append(neighborhoods.iloc[i,0])
# n_names =neighborhoods[['Neighborhood']]
n_names
```

```python
#isolate units occupied and rented
rented = neighborhoods[['Num_Occupied','Units_Rented']]
rented
```

```python
#calculate the percent of units rented in a neighborhood
units_rented =[]
for i in range(len(rented)) : 
    perct_rented =((rented.iloc[i, 1]) /(rented.iloc[i, 0])*100)
    rounded = perct_rented.round(2)
    units_rented.append(rounded)
   
print(len(units_rented)) 
```
